# Bacteria Growth 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nsteps = 100  #Number of steps in our lil simulation 
x = np.zeros(nsteps)  #Number of bacteria at every step 
x[0] = 1              #The Original number fo bacteria 
r = 1.1               #Rate of population change (babies - death) 

for i in range (1,nsteps):
    x[i] = x[i-1] *r
    
plt.plot(x);
    
```

```python
nsteps = 100  #Number of steps in our lil simulation 
x = np.zeros(nsteps)  #Number of bacteria at every step 
x[0] = 0.6              #The Original population size  
r = 3.44            #Rate of population change (babies - death) 

for i in range (1,nsteps):
    x[i] = r*x[i-1] *(1-x[i-1])
    
    
plt.plot(x);
plt.ylim(0,1); #Force y-axis to be between 0 and 1
```
