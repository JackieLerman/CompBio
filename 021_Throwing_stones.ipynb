# Simple Numerical Integration For Motion

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = -1       #acceleration; negative cuz it is going down towards earth surface 
v = 5       #We throw the stone upwards so it is positive
x = 0       #We throw it from the surface of earth, so x =0
dt = 0.01   # Our numerical integration step is supposed to be small 
nsteps = 1000

history = np.zeros(nsteps) # we will plot this eventually

for i in range(nsteps):
    v = v + a*dt
    x = max(x + v*dt,0)
    history[i]= x
    
plt.plot(history)
```

```python
a = -1       #acceleration; negative cuz it is going down towards earth surface 
vy = 5       #We throw vertical the stone upwards so it is positive
vx = 1       # horizontal component of speed 
y = 0       #Throwing from our house 
x = 0       #We throw it from the surface of earth, so x =0
dt = 0.1   # Our numerical integration step is supposed to be small 
f = .1    #How important friction is relative to gravity 
nsteps = round(10/dt)



for i in range(nsteps):
    vy = vy + a*dt - vy *f
    vx = vx - vx*f
    x = x + vx*dt    #Update xpos
    y = max(y + vy*dt,0)  #update ypos
    plt.plot(x,y,"k.")
    
```
