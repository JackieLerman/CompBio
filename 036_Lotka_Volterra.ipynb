```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nsteps = 90000
a = 1 
b = 1 
c = 0.5 
d = 0.5 

dt = 0.001 

x = np.zeros(nsteps)
y = np.zeros(nsteps)

x[0] = 1
y[0] = .2

for t in range (1,nsteps):
    x[t] = x[t-1] + dt*(a*x[t-1] - b*x[t-1]*y[t-1])
    y[t] = y[t-1] + dt*(c*x[t-1]*y[t-1] - d*y[t-1])
    
    
    
fig = plt.figure();
plt.plot(x, label = 'bunnies')
plt.plot(y,label = 'Foxes')
plt.legend(loc= "upper right")

```
