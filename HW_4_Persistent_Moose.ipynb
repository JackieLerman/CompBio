```python
import numpy as np
import time 
import matplotlib.pyplot as plt 
```

```python
%matplotlib notebook 

#Magic Spell ^^

n_animals = 100
n_steps = 1000
difspeed = 0.01   #Diffusion speed.

vx = np.zeros(n_animals)
vy = np.zeros(n_animals)

x = np.ones(n_animals)*0.5
y = np.ones(n_animals)*0.5 
    
#Plotting 
fig = plt.figure(figsize = (4,4));
plt.xlim((0,1))
plt.ylim((0,1))


for s in range(n_animals): #Loop to set speed values for each moose, outside of other loop to ensure straight direction
    vx[s] = np.random.uniform()
    vy[s] = np.random.uniform()


for t in range(n_steps):
    
    x += vx
    y += vy
        
    #Reversed all ones and zeros to create pacman/donut world effect...
    x[x<0] = 1
    x[x>1] = 0
    y[y<0] = 1
    y[y>1] = 0
       
    plt.cla()
    plt.plot (x,y,'k.',markersize = 2);
    plt.xlim((0,1))
    plt.ylim((0,1))
    fig.canvas.draw()
    
    
   

```
