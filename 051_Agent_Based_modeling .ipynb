# Deer Diffusion

```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
%matplotlib notebook 
#Magic Spell 
n_animals = 100
n_steps = 1000
difspeed = 0.01   #Diffusion speed.  

x = np.ones(n_animals)*0.5
y = np.ones(n_animals)*0.5 
    
#Plotting 
fig = plt.figure(figsize = (4,4));
plt.xlim((0,1))
plt.ylim((0,1))




for t in range(n_steps):
    x += np.random.randn(n_animals) *difspeed
    y += np.random.randn(n_animals) *difspeed
    
    for d in range(n_animals):
        if x[d] >1:
            x[d] = 1
            
        if x[d] < 0:
            x[d] = 0
            
        if y[d] >1:
            y[d] = 1
            
        if y[d] < 0:
            y[d] = 0
        
        
    plt.cla()
    plt.plot (x,y,'k.',markersize = 2);
    plt.xlim((0,1))
    plt.ylim((0,1))
    fig.canvas.draw()
    
    
   


```
